```python
# new dataframe of selected counties
counties = pd.read_csv('county_info.csv', index_col = [0])
counties.rename({'counties': 'county'}, axis='columns', inplace = True)
counties
```

```python
states = []
counties_lst = []

for index, county in counties.iterrows():
    states.append(county['county'][-2:])
    counties_lst.append(county['county'][0:-3])
    
counties['state'] = states
counties['county'] = counties_lst


counties = counties[['county', 'state', 'populations', 'ed_rate', 'tier']]

#counties
```

```python
c19df = pd.DataFrame()

for index, county in counties.iterrows():
    c19df = c19df.append(c19cases[(c19cases['Province_State'] == str(county['state'])) & 
                          (c19cases['Admin2'] == str(county['county']))])
    
    

c19df

#c19df.to_csv('covid_cases.csv')
```
